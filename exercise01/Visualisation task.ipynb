```python
import pandas as pd
import random
import numpy as np
import seaborn as sns
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

<h1>A New study says that exercise is associated to increased levels of cholesterol.</h1>
    
Researchers analised the data of 224 patients to investigate the relation between Exercise and cholesterol levels.

```python
#Read the dataset
df = pd.read_csv('./dataset.csv')
#Display the first 15 rows of the dataset
df.head(15)
```

<h1>First the researchers visualised the data using a scatter plot</h1>

```python
g = sns.scatterplot(x="Exercise", y="Cholesterol",data=df, s=70)

#set the limits of the plot
#plt.xlim(0,100)
#plt.ylim(0,120)

#set labels size
plt.xlabel('Exercise level', fontsize=18, fontweight='bold')
plt.ylabel('Cholesterol', fontsize=18, fontweight='bold')

#set ticks size
plt.xticks(fontsize=18)
t = plt.yticks(fontsize=18)
```

<h1>Then they ran a linear regression model and got to the conclusion that more exercise is associated to higher levels of cholesterol</h1>

High correlation: 0.83

```python
#plot the data and fit a linear regression line
p = sns.lmplot(x="Exercise", y="Cholesterol",data=df)

#compute the correlation
corr = pearsonr(df['Exercise'],df['Cholesterol'])
plt.text(100, 200, "Correlation: {0:.2f}".format(corr[0]), fontsize= 18)

#set labels size
plt.xlabel('Exercise level', fontsize=18, fontweight='bold')
plt.ylabel('Cholesterol', fontsize=18, fontweight='bold')

#set ticks size
plt.xticks(fontsize=18)
t = plt.yticks(fontsize=18)
```
